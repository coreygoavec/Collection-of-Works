```r
# Get the minimum, maximum, and average of the "Age" numeric variable compared to the "Marital Status" categorical variable.
# install.packages("dplyr")
library("dplyr")

# Creating a mew table
by_status <- risk_data %>%
  group_by(risk_data$`Marital Status`) %>%
  summarize(nobs = n(),
            nobs_pct = nobs/sum(nobs),
            avg_age = mean(Age, na.rm = TRUE),
            min_age = min(Age),
            max_age = max(Age))

by_status
```

```r
# Conduct a bootstrapped sampling of the mean difference of"Savings" across "Credit Risk"
# install.packages("mosaic")
library("mosaic")

# Doing a single bootsrapp
risk_boot <- mosaic::resample(risk_data)

# Getting "Savings" across "Credit Risk"
print(mean(Savings ~ `Credit Risk`, data = risk_data))

# Getting the difference in the mean
cat("Mean difference of bootstrapped sample:", diffmean(Savings ~ `Credit Risk`, data = risk_boot))
cat("\nThe savings do differ among the two groups, the mean savings between those who do not have a credit risk are higher than those who are at a credit risk.")
```

```r
# Conduct a bootstrapped technique with a sample size of 10,000.
risk_bt10k <- do(10000) * diffmean(Savings ~ `Credit Risk`, data = mosaic::resample(risk_data))
```

```r
# Histogram of the mean difference.
hist(risk_bt10k$diffmean, col = 'violet', main = "Bootstrapped Diff Mean")
```

```r
# Standard error and 99% confidence interval
# Standard error // used with 10000 bootstrapp sample
risk_bt10k %>%
  summarize(Savings_StdErr = sd(diffmean))

# Confidence interval
print(confint(risk_bt10k, level = 0.99))
```
